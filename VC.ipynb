### this cell contains a function to get the information of videos

#### inputs that are our points in this cell are video codec, resolution, frame rate, aspect ratio, video bitrate(bitrate), duration, and so on; and as you suggested these are not required specifications but details of videos

```python
def get_video_info(file_path):
    cmd = ["ffprobe", "-v", "error", "-show_format", "-show_streams", "-print_format", "json", file_path]
    result = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output, error = result.communicate()
    if error:
        raise Exception(f"Error occurred: {error.decode('utf-8')}")

    return json.loads(output)
```

## Finally this cell converts our files to the files with -formatOK and adjusts attributes of each file the convertion of the files is done based on the attributes of each file; and all of the proccess is on the field of the documentation

```python
for root, _, files in os.walk(extracted_dir):
    for file in files:
        if file.lower().endswith(".avi") or file.lower().endswith(".mov") or file.lower().endswith(".mp4"):
            input_path = os.path.join(root, file)
            output_path = os.path.join(root, os.path.splitext(file)[0] + "_formatOK.mp4")

            # Use ffmpeg to convert the film
            cmd = [
                "ffmpeg",
                "-i", input_path,
                "-c:v", "libx264",
                "-c:a", "aac",
                "-r", "25",
                "-s", "640x360",
                "-b:v", "2M",  # Adjust as needed
                "-b:a", "256k",
                "-strict", "experimental",
                output_path
            ]

            subprocess.run(cmd)

            print(f"Converted {file} to {output_path}")
```
